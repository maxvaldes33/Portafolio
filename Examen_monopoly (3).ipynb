## Importa el archivo excel y las librerias a utilizar

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
archivo_excel = "C:\\Users\\musrr\\Downloads\\Base_clientes_Monopoly.xlsx"
dataframe = pd.read_excel(archivo_excel)
print(dataframe.head())
```

## Crear un histograma con valores de Renta

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=dataframe, x='Renta', bins=40, kde=True, color='skyblue', alpha=0.7)
plt.xlabel('Distribucion de la Renta')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Renta')
plt.legend(labels=['Distribuci√≥n', 'KDE'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajusta las etiquetas del eje x
renta_ticks = [i * 1000000 for i in range(11)]
plt.xticks(renta_ticks, rotation=0)
plt.show()
```

# Fase 3: Preparacion de los datos, Data Preparing

**Contamos la cantidad de nulos en la tabla de Renta**

```python
nulos_renta = dataframe['Renta'].isnull().sum()
print("Cantidad de valores nulos en 'Renta':", nulos_renta)
```

**Le remplazamos los valores null por el promedio de la Renta**

```python
dataframe['Renta'].fillna(dataframe['Renta'].mean(), inplace=True)
```

**Volvemos a contar la cantidad de nulos en la tabla Renta**

```python
nulos_renta = dataframe['Renta'].isnull().sum()
print("Cantidad de valores nulos en 'Renta':", nulos_renta)
```

Se rellenan los datos faltantes de renta con el promedio (mean), en la tabla a continuacion se puede comparar con la anterior:

```python
estadisticas_renta = dataframe['Renta'].describe()
print(estadisticas_renta)
```

## La nueva distribucion de la renta

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=dataframe, x='Renta', bins=40, kde=True, color='skyblue', alpha=0.7)
plt.xlabel('Distribucion de la Renta')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Renta')
plt.legend(labels=['Distribuci√≥n', 'KDE'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajusta las etiquetas del eje x
renta_ticks = [i * 1000000 for i in range(11)]
plt.xticks(renta_ticks, rotation=0)
plt.show()
```

Ac√° se puede apreciar como el punto mas alto de la renta cambio, ya que la media relleno a los valores faltantes, la ventaja de rellenar con la media

## Ahora cambiaremos los valores nulos de la tabla PIN por 0

Con el siguiente codigo xontaremos los nulos y despues los remplazaremos a los valores faltantes en la tabla CambioPin por 0, ya que existian registros solo si los usuarios habian cambiado su pin en al menos una ocasi√≥n, de no ser as√≠, era null

```python
nulos_pin = dataframe['CambioPin'].isnull().sum()
print("Cantidad de valores nulos en 'CambioPin':", nulos_pin)
```

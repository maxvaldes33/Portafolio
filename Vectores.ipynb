```python
import matplotlib.pyplot as plt
import numpy as np
import sympy
```

```python
# imprimir con notación matemática.
sympy.init_printing(use_latex='mathjax')
```

```python
# graficando vector en R^2 [2, 4]
def dibujar_figura():
    """Crea la figura de pyplot y los ejes."""
    fix, ax = plt.subplots()
    for spine in ["left", "bottom"]:
        ax.spines[spine].set_position("zero")

    return ax
```

```python
def vector_plano(vector, color):
    """Genera el grafico de los vectores en el plano"""
    v = vector
    ax.annotate(" ", xy=v, xytext=[0, 0], color=color,
                arrowprops=dict(facecolor=color,
                                shrink=0,
                                alpha=0.7,
                                width=0.5))
    ax.text(1.1 * v[0], 1.1 * v[1], v)
```

```python
ax = dibujar_figura()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.grid()
vector_plano([2, 4], "blue")

# Suma de dos vectores
# [3, 2] + [1, -2]

ax = dibujar_figura()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.grid()

vecs = [[3, 2], [1, -2]]

for v in vecs:
    vector_plano(v, "blue")
v = np.array([2, 4]) + np.array([2, -2])
vector_plano(v, "red")

# Multiplicacion de un vector con un escalar
# [1, 2] * 2
ax = dibujar_figura()
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.grid()

v = np.array([1, 2])
vector_plano(v, "blue")

v = v * 2
vector_plano(v, "red")
```

# Resolviendo la imagen 1

```python
# Coeficientes de las ecuaciones
coeficientes = np.array([[1, 5], [-2, -7]])
constantes = np.array([7, -5])

# Resolver el sistema de ecuaciones
solucion = np.linalg.solve(coeficientes, constantes)

# Obtener los valores de x1 y x2
x1 = solucion[0]
x2 = solucion[1]

print("Solución:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
```

# Resolviendo Imagen 2

```python
# Coeficientes de las ecuaciones
coeficiente = np.array([[2, 4], [5, 7]])
constantes = np.array([-4, 11])

# Resolver el sistema de ecuaciones
# La función np.linalg.solve resuelve el sistema de ecuaciones y devuelve los valores de x1 y x2.
solucion = np.linalg.solve(coeficiente, constantes)

# Obtener los valores de x1 y x2
x1 = solucion[0]
x2 = solucion[1]

print("Solución:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
```

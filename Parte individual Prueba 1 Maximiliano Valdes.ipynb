```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\musrr\\Downloads\\Variables.csv", delimiter=';')
```

Luego de importar pandas y matplotlib.pyplot para poder realizar graficos, agregamos el documento .csv, en el cual modificamos la base de datos puesto que esta separada por ';', y a continuacion mapeamos el sexo de 1 de mujer y 2 de hombre para que tengan sus valores dados por el diccionarios, y hacemos lo mismo con tamaño, categoria y grupo, para luego aplicar los cambios y mostrar los primeros 5 datos con un .head para así ver si se realizaron los cambios

```python
#Definir los numeros y letras a sus respectivas categorias
mapeo_sexo = {1: 'Mujer', 2: 'Hombre'}
mapeo_tamano = {1: 'Pequeño', 2: 'Mediana', 3: 'Grande'}
mapeo_categoria = {
    'B': 'Minería',
    'C': 'Industria Manufacturera',
    'D': 'Suministro de Electricidad, Gas, Vapor y Aire Acondicionado',
    'E': 'Suministro de Agua, Evacuación de Aguas Residuales, Gestión de Desechos y Descontaminación',
    'F': 'Construcción',
    'G': 'Comercio',
    'H': 'Transporte y Almacenamiento',
    'I': 'Actividades de Alojamiento y Servicio de Comidas',
    'J': 'Información y Comunicaciones',
    'K': 'Actividades Financieras y de Seguros',
    'L': 'Actividades Inmobiliarias',
    'M': 'Actividades Profesionales, Científicas y Técnicas',
    'N': 'Actividades de Servicios Administrativos y de Apoyo',
    'O': 'Administración Pública',
    'P': 'Enseñanza',
    'Q': 'Actividades de Atención de la Salud Humana y de Asistencia Social',
    'R': 'Actividades Artísticas, de Entretenimiento y Recreativas'}
mapeo_grupo = {
    1: 'Directores y gerentes',
    2: 'Profesionales',
    3: 'Técnicos',
    4: 'Trabajadores de apoyo administrativo',
    5: 'Trabajadores de servicios personales, protección y seguridad',
    6: 'Vendedores, promotores y modelos',
    7: 'Operarios manuales y artesanos',
    8: 'Operadores y montadores de instalaciones y máquinas',
    9: 'Trabajadores no especializados'}

# Aplicar los cambios a sus respectivas categorias
df['sexo'] = df['sexo'].replace(mapeo_sexo)
df['tamano'] = df['tamano'].replace(mapeo_tamano)
df['categoria'] = df['categoria'].replace(mapeo_categoria)
df['grupo'] = df['grupo'].replace(mapeo_grupo)
# Mostrar el df con un .head()
df.head()
```

A continuacion se convierte los valores no numericos de la columna 'ro_t_1' y si posee alguna clase de error se transforman a valor NaN

```python
df['ro_t_1'] = pd.to_numeric(df['ro_t_1'], errors='coerce')
```

A continuacion se eliminan los valores nulos del target

```python
df = df.dropna(subset=['ro_t_1'])
```

Para no tener problemas de interpretacion transforme los valores elevados a numeros con 2 decimales, luego agrupo por sexo el target y presento los valores estadisticos de la misma.

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

Aquí muestro las medidas de tendencia central con respecto a la columna 'ro_t_1' la cual es la remuneracion y la vemos con los valores de las categorias, los separo por sexo y su respectivo grupo, por desgracia son tantos valores que llega a ser practicamente imposible mostrarlos todos en un mismo grupo, por lo mismo en los siguientes codigos los separamos de distintas agrupaciones.

```python
resultados = df.groupby(['categoria', 'sexo', 'grupo']).agg({
    'ro_t_1': ['mean', 'median', 'std', 'min', 'max']
}).reset_index()
resultados.head(300)
```

```python
grupos_por_sexo = df.groupby('sexo')['ro_t_1']

estadisticos_sexo = grupos_por_sexo.describe()

estadisticos_sexo.head()
```

Agrupamos el target ahora con los grupos y calculamos sus estadisticas descriptivas de manera individual

```python
grupos_por_grupo = df.groupby('grupo')['ro_t_1']
estadisticos_grupo = grupos_por_grupo.describe()

estadisticos_grupo.head(9)
```

Agrupo el target ahora con las categorias y calculamos sus estadisticas descriptivas de manera individual

```python
grupos_por_categoria = df.groupby('categoria')['ro_t_1']
estadisticos_categoria = grupos_por_categoria.describe()
estadisticos_categoria.head(17)
```

Agrupo el grupo con el sexo para diferenciar los grupos con sus respectivos sexos.

```python
grupos_por_grupo_y_sexo = df.groupby(['grupo', 'sexo'])['ro_t_1']
estadisticos_grupo_sexo = grupos_por_grupo_y_sexo.describe()
estadisticos_grupo_sexo.head(20) 
```

Ahora agrupamos las categorias con los grupos y como son tantas categorias en los distintos grupos estas no se pueden ver en su totalidad

```python
grupos_por_grupo_y_categoria = df.groupby(['grupo', 'categoria'])['ro_t_1']
estadisticos_grupo_categoria = grupos_por_grupo_y_categoria.describe()
estadisticos_grupo_categoria.head(151) 
```

Y como ultima estadistica esta categoria con sexo

```python
grupos_por_categoria_sexo = df.groupby(['categoria', 'sexo'])['ro_t_1']
estadisticos_categoria_sexo = grupos_por_categoria_sexo.describe()
estadisticos_categoria_sexo.head(300)
```

Como conclusión podemos notar que muchos mas hombres posee remuneración en comparación a las mujeres, puesto que hay casí 9000 hombres y menos casí 7600 mujeres, lo cual nos muestra que en gran parte hay muchos mas hombres con trabajo que mujeres, por otra parte se puede ver que la remuneración minima en hombres es menor a la de las mujeres, pero la remuneracion maxima en los hombres es bastante mayor, con casí 2 millones de diferencia.

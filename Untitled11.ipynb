```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
    name=fn
df = pd.read_csv(name, sep=",")
df.head(10)
```

```python
columnas_seleccionadas = [ 'Transaction ID', 'Order ID', 'Customer ID', 'Product Name', 'Sub-Category', 'Category', 'Discount', 'Order Date', 'Order Day', 'Order Month', 'Order Weekday' ,'Order Year' ,'Quantity','Segment', 'State', 'Profit CLP', 'Sales CLP']
dataframe_seleccionado = df[columnas_seleccionadas]
correlation_matrix = dataframe_seleccionado.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Mapa de Correlaci√≥n')
plt.show()
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="openai-community/gpt2")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("dccuchile/gpt2-spanish")
model = AutoModelForCausalLM.from_pretrained("dccuchile/gpt2-spanish")
```

```python
# prompt: crea un codigo para poder generar texto

text = "Eran las seis y media y las manecillas seguían tranquilamente hacia delante, ya había pasado incluso la media, eran ya casi las menos cuarto. «¿Es que no habría sonado el despertador?» Desde la cama se veía que estaba correctamente puesto a las cuatro, seguro que también había sonado. Sí, pero... ¿era posible seguir durmiendo tan tranquilo con ese ruido que hacía temblar los muebles? Bueno, tampoco había dormido tranquilo, pero quizá tanto más profundamente."

# Use the pipeline to generate text
output = pipe(text, num_return_sequences=3, max_new_tokens=100) # limit number of return sequences and set max_new_tokens

# Print the generated text
print(output[0]["generated_text"])
print(output[1]["generated_text"])

# Use the model directly to generate text
inputs = tokenizer(text, return_tensors="pt")
output = model.generate(**inputs, max_length=264) # set max_length to length of input text + 100

# Decode the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

# Print the generated text
print(generated_text)
```
